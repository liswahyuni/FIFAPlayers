For FIFA17 - FIFA22 "Jersey Number" column is equal to FIFA23 "Kit Number" column.

“Release Clause” atau “Buy-Out Clause” adalah klausul dalam kontrak yang menetapkan biaya transfer minimum seorang pemain, yang telah disepakati di kontrak sang pemain dan klub yang bersangkutan. Jika satu tim melayangkan tawaran yang sama dengan release clause sang pemain, maka klub pemilik pemain tersebut harus menerima tawaran tersebut dan mempersilakan pemainnya untuk berbicara dengan klub penawar. Release clause ini berbagai macam bentuk klausulnya, bergantung kepada kesepakatan yang terjadi antara pemain dan klub saat pembicaraan kontrak.

```python
def convert_currency(val):
    if isinstance(val, str):
        new_val = val.replace(',', '').replace('€', '')
        if new_val[-1] == 'B':
            return float(new_val[:-1]) * 1000000000
        elif new_val[-1] == 'M':
            return float(new_val[:-1]) * 1000000
        elif new_val[-1] == 'K':
            return float(new_val[:-1]) * 1000
        else:
            return float(new_val)
    else:
        return val

cols = ['Value', 'Wage', 'Release Clause']
data_concat[cols] = data_concat[cols].applymap(convert_currency)
```

```python
data_concat["Jersey Number"] = data_concat.apply(lambda x: x['Kit Number'] if pd.isnull(x['Jersey Number']) else x['Jersey Number'], axis = 1)
data_concat.drop("Kit Number", axis=1, inplace=True)
```

Dalam data FIFA player, “Position” mengacu pada posisi utama pemain di lapangan, sedangkan “Best Position” mengacu pada posisi di mana pemain memiliki peringkat keseluruhan tertinggi. Sebagai contoh, seorang pemain mungkin memiliki “Position” sebagai gelandang tengah (CM), tetapi memiliki peringkat keseluruhan tertinggi ketika bermain sebagai gelandang bertahan (CDM), sehingga “Best Position” mereka adalah CDM.

```python
# Define function to get characters after '>'
def get_char_after_gt(x):
    if '>' in x:
        return x.split('>')[1]
    else:
        return x
```

```python
data_concat['Position'] = data_concat['Position'].apply(get_char_after_gt)
```
